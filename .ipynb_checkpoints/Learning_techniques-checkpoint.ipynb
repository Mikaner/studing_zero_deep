```python
import sys, os
sys.path.append(os.pardir)
import numpy as np
from common.two_layer_net import TwoLayerNet
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from collections import OrderedDict
from common.optimizer import *
```

```python
class SGD:
    def  __init__(self, lr=0.01):
        self.lr = lr
        
    def update(self, params, grads):
        for key in params.keys():
            params[key] -= self.lr * grads[key]
```

```python
def f(x,y):
    return (1/20)*(x**2) + y**2
def df_dw0(w0, w1):
    return 1/10 * w0
def df_dw1(w0, w1):
    return 2*w1
```

```python
x = np.arange(-10,10,0.1)
y = np.arange(-10,10,0.1)

X,Y = np.meshgrid(x, y)
Z = f(X,Y)

fig = plt.figure()
ax = Axes3D(fig)

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")

ax.plot_wireframe(X, Y, Z)
plt.show()
```

```python
# 
w_range = 10
dw = 1
w0 = np.arange(-w_range, w_range, dw)
w1 = np.arange(-w_range, w_range, dw)
wn = w0.shape[0]

dw0 = np.zeros((len(w0), len(w1)))
dw1 = np.zeros((len(w0), len(w1)))

for i0 in range(wn):
    for i1 in range(wn):
        dw0[i1,i0] = df_dw0(w0[i0], w1[i1])
        dw1[i1, i0] = df_dw1(w0[i0], w1[i1])

plt.quiver(w0, w1, dw0, dw1)
plt.show()
```

```python
class Momentum:
    def __init__(self, lr=0.01, momentum=0.9):
        self.lr = lr
        self.momentum = momentum
        self.v = None
        
    def update(self, params, grads):
        if self.v is None:
            self.v = {}
            for key, val in params.items():
                self.v[key] = np.zeros_like(val)
                
            for key in  params.keys():
                self.v[key] = self.momentum*self.v[key] - self.lr*grads[key]
                params += self.v[key]
```

```python
class AdaGrad:
    def __init__(self, lr=0.01):
        self.lr = lr
        self.h = None
        
    def update(self, params, grads):
        if self.h is None:
            self.h = {}
            for key, val in params.items():
                self.h[key] = np.zeros_like(val)
                
            for key in params_keys():
                self.h[key] += grads[key] * grads[key]
                params[key] -= self.lr * grads[key] / (np.sqrt(self.h[key]) + 1e-7)
```
